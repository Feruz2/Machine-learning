```
import pandas as pd

df = pd.read_csv('imdb_top_1000.csv')

average= df['IMDB_Rating'].mean()
genres_count = df['Genre'].nunique()
column = df.columns.tolist()
results_df = pd.DataFrame({'Average Rating': [average], 'Unique Genres Count': [genres_count], 'Column Names': [column]})
results_df.to_csv('results.csv', index=False)
```

```
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('imdb_top_1000.csv')
df['Genre'] = df['Genre'].str.split(', ')
flat_genre_list = [genre for genres in df['Genre'] for genre in genres]
genre_list_to_series = pd.Series(flat_genre_list)
genre_counts = genre_list_to_series.value_counts()



plt.figure(figsize=(10, 6))
plt.bar(genre_counts.index, genre_counts.values)
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Genre Distribution of Movies')
plt.xticks(rotation=45)
plt.show()


print(genre_counts)
```

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('imdb_top_1000.csv')

movies_per_year = df['Released_Year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.scatter(movies_per_year.index, movies_per_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies in IMDb Top 1000 List per Year')
plt.xticks(rotation=45)
plt.show()
```

*Note*: The visualization below is done using Decade, as there was visual clutter on the x-axis when down against year.


```
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('imdb_top_1000.csv')

str_to_int = [int(year)//10 * 10 for year in df['Released_Year'] if year.isdigit()]

movies_per_decade = pd.Series(str_to_int).value_counts().sort_index()


plt.figure(figsize=(10, 6))
plt.scatter(movies_per_decade.index, movies_per_decade.values)
plt.xlabel('Decade')
plt.ylabel('Number of Movies')
plt.title('Number of Movies in IMDb Top 1000 List per Decade')
plt.xticks(rotation=45)
plt.show()
```
